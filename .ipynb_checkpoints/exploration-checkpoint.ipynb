Многие этничности получили всего несколько кодирований. **Возможно их стоит удалить?**

```python
eth_less_10_group_to_code = df["eth_group_to_code"].value_counts()[
    (df["eth_group_to_code"].value_counts() <= 10)
].index.tolist()

eth_less_10_seed_eth_group = df["seed_eth_group"].value_counts()[
    (df["seed_eth_group"].value_counts() <= 10)
].index.tolist()

eth_little_labels = eth_less_10_group_to_code + eth_less_10_seed_eth_group
```

```python
len(eth_little_labels), eth_little_labels
```

73 этничности закодированы менее 10 раз так или иначе. Исключим их.

```python
df.shape
```

```python
df = df[~df["eth_group_to_code"].isin(eth_less_10_group_to_code)]
df = df[~df["seed_eth_group"].isin(eth_less_10_seed_eth_group)]
```

Мы хотим выбрать документы, в которых хотя бы два человека ответили, что по отношению хотя бы к одной из этничностей в тексте присутствуют явные или скрытые призивы к насилию.

```python
a = df[["document.id", "assessor", "eth_group_to_code", "encourage_aggression_meaning_label"]]\
    .groupby(["document.id", "assessor", "eth_group_to_code"])\
    .first()\
    .agg({"encourage_aggression_meaning_label": lambda x: "yes" in x.lower() if type(x) is str else False})
a
```

```python
ids_of_encourage_aggression_posts = a.groupby(["document.id", "eth_group_to_code"]).sum()\
    .query("encourage_aggression_meaning_label == 2")\
    .index.get_level_values("document.id").unique().tolist()
```

```python
len(ids_of_encourage_aggression_posts)
```

```python
writer = pd.ExcelWriter("../output/encourage_aggression_posts.xlsx")
df[["document.id", "source_text"]]\
    [df["document.id"].isin(ids_of_encourage_aggression_posts)]\
    .drop_duplicates(subset="document.id").to_excel(writer, "texts_only", index=False)
```

```python
df[["document.id", "eth_group_to_code", "encourage_aggression_meaning_label", "source_text"]]\
    [df["document.id"].isin(ids_of_encourage_aggression_posts)].to_excel(writer, "full_output", index=False)
writer.save()
```
