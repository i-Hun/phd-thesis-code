```r
library("feather")
library("lme4")
library("lmerTest")
library("ggplot2")
library("sjPlot")
require("MuMIn")
library("xlsx")
```

```r
df <-  read_feather("/Users/hun/Google Drive/HSE/PhD/PhD thesis/data/df_regr.feather")
```

```r
names(df)
```

```r
df$document.id = as.factor(df$document.id)
df$document.id = as.factor(df$document.id)
df$represent_ethicity_meaning_value = as.factor(df$represent_ethicity_meaning_value)
df$eth_group_to_code = as.factor(df$eth_group_to_code)
df$assessor = as.factor(df$assessor)
df$aggregated_groups = as.factor(df$aggregated_groups)
```

# Зависимость отношения от этничности автора

```r
model = lmer(opinion_about_ethnonym_recoded_value ~ 
               represent_ethicity_meaning_value +
               (represent_ethicity_meaning_value|eth_group_to_code) +
               (represent_ethicity_meaning_value|assessor) +
               (represent_ethicity_meaning_value|document.id),
             data=df, REML = F)
```

```r
summary(model)
```

```r
r.squaredGLMM(model)
```

# Отношение в зависимости от этнической группы

```r
model_sent_eth = lmer(opinion_about_ethnonym_recoded_value ~ 
               eth_group_to_code +
               (1|document.id) +
                (1|assessor),
             data=df, REML = F)
```

```r
summary(model_sent_eth)
```

```r
r.squaredGLMM(model_sent_eth)
```

```r
model_sent_eth2 = lmer(opinion_about_ethnonym_recoded_value ~ 
               eth_group_to_code +
               (1|document.id),
             data=df, REML = F)
```

```r
summary(model_sent_eth2)
```

```r
r.squaredGLMM(model_sent_eth2)
```

В зависимости от принадлежности автора и этнической группы.

```r
model_eth_auth = lmer(opinion_about_ethnonym_recoded_value ~ 
               represent_ethicity_meaning_value +
               eth_group_to_code +
               (represent_ethicity_meaning_value|assessor) +
               (represent_ethicity_meaning_value|document.id),
             data=df, REML = F)
```

```r
summary(model_eth_auth)
```

```r
r.squaredGLMM(model_eth_auth)
```

```r
model_reg = lmer(opinion_about_ethnonym_recoded_value ~ 
                represent_ethicity_meaning_value +
                eth_group_to_code +
                (eth_group_to_code|region) +
                (1|assessor) +
                (1|document.id),
            data=df, REML = F)
```

```r
summary(model_reg)
```

```r
r.squaredGLMM(model_reg)
```

# Группировка этничностей

```r
model_sent_eth = lmer(opinion_about_ethnonym_recoded_value ~ 
               aggregated_groups +
               (1|document.id) +
                (1|assessor),
             data=df, REML = F)
```

```r
summary(model_sent_eth)
```

```r
(res_table <- as.data.frame(coef(summary(model_sent_eth))))
```

```r
model_sent_eth2 = lmer(opinion_about_ethnonym_recoded_value ~ 
               aggregated_groups *
                represent_ethicity_meaning_value +
               (1|document.id) +
                (1|assessor),
             data=df, REML = F)
```

```r
summary(model_sent_eth2)
```

```r
(res_table2 <- as.data.frame(coef(summary(model_sent_eth2))))
```
